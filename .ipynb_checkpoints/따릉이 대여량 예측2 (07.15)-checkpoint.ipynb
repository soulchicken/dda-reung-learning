```python
#이상치를 제거하는 함수 만들기
def removeOutliers(x, column):
    # Q1, Q3구하기
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    
    # 1.5 * IQR(Q3 - Q1)
    iqr = 1.5 * (q3 - q1)
    
    # 이상치를 제거
    y=x[(x[column] < (q3 + iqr)) & (x[column] > (q1 - iqr))]
    
    return(y)

#이상치를 바꾸는 함수 만들기
def changeOutliers(data, column):
    x=data.copy()
    # Q1, Q3구하기
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    
    # 1.5 * IQR(Q3 - Q1)
    iqr = 1.5 * (q3 - q1)
    
    #이상치 대체값 설정하기
    Min = 0
    if (q1 - iqr) > 0 : Min=(q1 - iqr)
        
    Max = q3 + iqr
    
    # 이상치를 변경
    # X의 값을 직졉 변경해도 되지만 Pyhon Warning이 나오기 떄문에 인덱스를 이용
    x.loc[(x[column] > Max), column]= Max
    x.loc[(x[column] < Min), column]= Min
    
    return(x)
```

```python
# 마안 약 지운다면?
train_remove=removeOutliers(train_df, 'hour_bef_temperature')
train_remove=removeOutliers(train_df, 'hour_bef_windspeed')
```

```python
# 마안 약 바꾼다면?
train_change=changeOutliers(train_df, 'hour_bef_temperature')
train_change=changeOutliers(train_df, 'hour_bef_windspeed')
```

## 3. 모델 학습 및 검증

```python
features = ['hour','hour_bef_temperature','hour_bef_windspeed']
X_train = train_remove[features]
y_train = train_remove['count']
X_test = test_df[features]
```

```python
X_train2 = train_change[features]
y_train2 = train_change['count']
X_test2 = test_df[features]
```

```python
from sklearn.ensemble import RandomForestRegressor

model100 = RandomForestRegressor(n_estimators=100,random_state=0)
model100_5 = RandomForestRegressor(n_estimators=100,max_depth=5,random_state=0)
model200 = RandomForestRegressor(n_estimators=200)
```

### 위에서 만든 모델 3개를 각각 fit을 이용해 학습을 해주고, predict을 통해 검증을 한다.

```python
model100.fit(X_train,y_train)
model100_5.fit(X_train,y_train)
model200.fit(X_train,y_train)

ypred1 = model100.predict(X_test)
ypred2 = model100_5.predict(X_test)
ypred3 = model200.predict(X_test)
```

```python
submission = pd.read_csv('data/submission.csv')
submission
```

```python
submission['count'] = ypred1
submission.to_csv('model102.csv',index=False)

submission['count'] = ypred2
submission.to_csv('model102_5.csv',index=False)

submission['count'] = ypred3
submission.to_csv('model202.csv',index=False)
```

```python
model100.fit(X_train2,y_train2)
model100_5.fit(X_train2,y_train2)
model200.fit(X_train2,y_train2)

ypred1 = model100.predict(X_test2)
ypred2 = model100_5.predict(X_test2)
ypred3 = model200.predict(X_test2)
```
